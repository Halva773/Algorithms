'''Задача 8. Создать класс Квадрат, который будет рисовать закрашенный квадрат,
зная координаты точек левого верхнего угла квадрата и  длину его стороны.
Координаты точки левого верхнего угла и длину стороны квадрата сделать приватными.'''

```python
class Square:
    def __init__(self, size, first_point: list):
        self.size = size
        self.__corner = first_point
        self.space = '-'
        
    def print(self, board_size=20):
        for i in range(self.__corner[0]):
            print(self.space*board_size)
        for i in range(self.size):
            print(self.space*self.__corner[1] + '*'*self.size + self.space*(board_size-self.size-self.__corner[1]))
        for i in range(board_size-self.size-self.__corner[0]):
            print(self.space*board_size)
            
a = Square(5, [1, 1])
a.print(board_size=7)
```
